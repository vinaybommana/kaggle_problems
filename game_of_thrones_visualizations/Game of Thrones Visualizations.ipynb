```python
def find_wins_of_attackers(df):
    attacker_wins = {}
    for king, outcome in zip(df.attacker_king, df.attacker_outcome):
        if not (king is np.nan or king != king) and not (outcome is np.nan or outcome != outcome):
            if outcome == "win":
                if king not in attacker_wins.keys():
                    attacker_wins[king] = 1
                else:
                    attacker_wins[king] += 1

    return attacker_wins

def find_wins_by_defenders(df):
    defender_wins = {}
    for king, outcome in zip(df.defender_king, df.attacker_outcome):
        if not (king is np.nan or king != king) and not (outcome is np.nan or outcome != outcome):
            if outcome == "loss":
                if king not in defender_wins.keys():
                    defender_wins[king] = 1
                else:
                    defender_wins[king] += 1
#         print(defender_wins)
    
    return defender_wins

def combine_wins(def_wins, attack_wins):
    combined_wins = {}
    for k, v in def_wins.items():
        if k not in combined_wins.keys():
            combined_wins[k] = v
        else:
            combined_wins[k] += v
            
    for k, v in attack_wins.items():
        if k not in combined_wins.keys():
            combined_wins[k] = v
        else:
            combined_wins[k] += v
    
    return combined_wins


attacker_wins = find_wins_of_attackers(df)
defender_wins = find_wins_by_defenders(df)
# print(defender_wins)
# print(attacker_wins)
combined_wins = combine_wins(defender_wins, attacker_wins)
wins_by_attacker = pd.DataFrame({'king': list(combined_wins.keys()), 'wins': list(combined_wins.values())})
wins_by_attacker
```

```python
sns.barplot(x=wins_by_attacker.king, y=wins_by_attacker.wins)
plt.xticks(rotation=45)
```
